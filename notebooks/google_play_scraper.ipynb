### App Detail

```python
from google_play_scraper import app

result = app(
    'com.nianticlabs.pokemongo',
    lang='en', # defaults to 'en'
    country='us' # defaults to 'us'
)
```

```python
print(result)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(result.items())
```

```python
df.T
```

```python
# Create a DataFrame
df = pd.DataFrame(result.items(), columns=["Key", "Value"])
```

```python
df.head()
```

```python
# Create a DataFrame from the dictionary
df = pd.DataFrame.from_dict(result, orient='index', columns=['Value'])
df.index.name = 'Key'
```

### App Reviews

```python
from google_play_scraper import Sort, reviews

result, continuation_token = reviews(
    'com.whatsapp',
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=1, # defaults to Sort.NEWEST
    count=3, # defaults to 100
    # filter_score_with=5, # defaults to None(means all score)
)

# If you pass `continuation_token` as an argument to the reviews function at this point,
# it will crawl the items after 3 review items.

result, _ = reviews(
    'com.fantome.penguinisle',
    continuation_token=continuation_token # defaults to None(load from the beginning)
)
```

```python
print(result)
```

```python
# Create a DataFrame from the dictionary
df = pd.DataFrame(result)
```

```python
df
```

```python
df.shape
```

### App All Reviews

```python
from google_play_scraper import Sort, reviews_all
app_id = 'in.evolve.android'
result = reviews_all(
    app_id,
    sleep_milliseconds=0, # defaults to 0
    lang='en', # defaults to 'en'
    country='us', # defaults to 'us'
    sort=Sort.NEWEST, # defaults to Sort.MOST_RELEVANT
    # filter_score_with=5 # defaults to None(means all score)
)
```

```python
print(result)
```

```python
df = pd.DataFrame(result)
df
```

### Save as CSV file

```python
import os
```

```python
# Specify the directory where you want to save the file
directory = 'data/'
```

```python
# Create the directory if it doesn't exist
if not os.path.exists(directory):
    os.makedirs(directory)
```

```python
# Save the DataFrame to a CSV file with the app_id name in the specified directory
csv_filename = os.path.join(directory, f'{app_id}.csv')
df.to_csv(csv_filename, index=False)
```
